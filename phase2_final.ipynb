```python
#  create time matrix using openrouteservice api
import requests
def send_request(locs, divider):
  body = {"locations":locs,"destinations": list(range(divider, len(locs))),"metrics":["duration"],"sources": list(range(0,divider))}
  headers = {
      'Accept': 'application/json, application/geo+json, application/gpx+xml, img/png; charset=utf-8',
      'Authorization': '5b3ce3597851110001cf624877b6decbd73c49d4844e69cfb6e884d9',
      'Content-Type': 'application/json; charset=utf-8'
  }
  call = requests.post('https://api.openrouteservice.org/v2/matrix/driving-hgv', json=body, headers=headers)
  matrix = call.json()['durations']
  return matrix


# build and combine matrices
main_matrix = [[0]*len(coords) for i in range(len(coords))]
rem = len(coords) % 50
# 
num_blocks = (int)((len(coords) - rem) / 50)
if (num_blocks == 0):
  main_matrix = send_request(coords + coords, len(coords))
else:

  for r in range(num_blocks):
    for c in range(num_blocks):
      dest = coords[c * 50 : c * 50 + 50]
      orig = coords[r * 50 : r * 50 + 50]
      dest = coords[c * 50 : c * 50 + 50]

      small_matrix = send_request(orig + dest, len(orig))
      # join matrices to main matrix
      for i in range (len(small_matrix)):
        for j in range (len(small_matrix)):
          main_matrix[r*50 + i][c*50 + j] = small_matrix[i][j]
      # edge cases
    if (rem > 0): 
      orig = coords[r*50 : r*50 +50]
      dest = coords[50*num_blocks : len(coords)]
      small_matrix = send_request(orig + dest, len(orig))
      # join matrices to main matrix
      for i in range (len(small_matrix)):
        for j in range (len(small_matrix[0])):
          main_matrix[r*50 + i][num_blocks*50 + j] = small_matrix[i][j]

  # bottom edge cases
  if (rem > 0):
    for j50 in range(num_blocks):
        orig = coords[num_blocks*50 : len(coords)]
        dest = coords[j50*50 : j50*50 + 50]
        small_matrix = send_request(orig + dest, len(orig))
        # join matrices to main matrix
        for i in range (len(small_matrix)):
          for j in range (len(small_matrix[0])):
            main_matrix[num_blocks*50 + i][j50*50 + j] = small_matrix[i][j]
        # edge 
        orig = coords[num_blocks*50 : len(coords)]
        dest = coords[50*num_blocks : len(coords)]
        small_matrix = send_request(orig + dest, len(orig))
        # join matrices to main matrix
        for i in range (len(small_matrix)):
          for j in range (len(small_matrix[0])):
            main_matrix[num_blocks*50 + i][num_blocks*50 + j] = small_matrix[i][j]
        # print(main_matrix)

banana_time_matrix = np.array(main_matrix)/60
print(len(banana_time_matrix),"x",len(banana_time_matrix[0]),"time matrix constructed, value = " + str(banana_time_matrix[0][1]))
```

```python
# shippers to hub solver
from functools import partial

from ortools.constraint_solver import pywrapcp
from ortools.constraint_solver import routing_enums_pb2

###########################
# Problem Data Definition #
###########################
def create_data_model():
    """Stores the data for the problem"""

    data = {}
    data['time_matrix'] = banana_time_matrix
    data['time_windows'] = time_windows
    data['demands'] = bananasZips['unit_count']
    data['time_per_demand_unit'] = 0 # 0 minutes/unit
    data['temperature'] = temperature_filter
#     TODO: change number of trucks
    data['num_16trucks'] = 10
    data['num_53trucks'] = 20
    data['num_vehicles'] = data['num_16trucks'] + data['num_53trucks']
    data['vehicle_capacities'] = [unit_capacity_16]*data['num_16trucks'] + [unit_capacity_53]*data['num_53trucks']
    data['depot'] = 0
    data['output'] = []
    return data


#######################
# Problem Constraints #
#######################

def create_demand_evaluator(data):
    """Creates callback to get demands at each location."""
    _demands = data['demands']

    def demand_evaluator(manager, node):
        """Returns the demand of the current node"""
        return _demands[manager.IndexToNode(node)]

    return demand_evaluator


def create_time_evaluator(data):
    def travel_time(data, from_node, to_node):
        """Gets the travel times between two locations."""
        if from_node == to_node:
            travel_time = 0
        else:
            travel_time = data['time_matrix'][from_node][to_node]
        return travel_time

    _total_time = data['time_matrix']
    def time_evaluator(manager, from_node, to_node):
        """Returns the total time between the two nodes"""
        return _total_time[manager.IndexToNode(from_node)][manager.IndexToNode(
            to_node)]
    return time_evaluator

# cost evaluator per vehicle
def create_cost_evaluator(data,vehindex):
    """Created call back to get total cost between locations """
    def travel_cost(data, from_node, to_node):
        if from_node == to_node:
            return 0
        else:
            hrs = (float)(data['time_matrix'][from_node][to_node] / 60)
            
            if (data['vehicle_capacities'][vehindex] == unit_capacity_16):
                rent1 = rent_16
                labor1 = labor_16
                mpg1 = mpg_16
                temp_cost1 = temp_info['temp_cost_16']
            elif (data['vehicle_capacities'][vehindex] == unit_capacity_53):
                rent1 = rent_53
                labor1 = labor_53
                mpg1 = mpg_53
                temp_cost1 = temp_info['temp_cost_53']
            cost = (rent1 + labor1 + temp_cost1) * hrs + gas_price * hrs * truck_speed/mpg1
            return cost
        
    _total_cost = [[0]*len(data['demands']) for i in range(len(data['demands']))]
    for from_node in range(len(data['demands'])):
        _total_cost[from_node] = {}
        for to_node in range(len(data['demands'])):
            if from_node == to_node:
                _total_cost[from_node][to_node] = 0
            else:
                _total_cost[from_node][to_node] = (travel_cost(data, from_node, to_node))
    def cost_evaluator(manager, from_node, to_node):
        return _total_cost[manager.IndexToNode(from_node)][manager.IndexToNode(
            to_node)]
    return cost_evaluator

# unit capacity constraints
def add_capacity_constraints(routing, data, demand_evaluator_index):
    """Adds capacity constraint"""
    capacity = 'Capacity'
    routing.AddDimensionWithVehicleCapacity(
        demand_evaluator_index,
        0,  # null capacity slack
        data['vehicle_capacities'],
        True,  # start cumul to zero
        capacity)

def add_time_window_constraints(routing, manager, data, time_evaluator_index):
    """Add Global Span constraint"""
    time = 'Time'
    horizon = 12000000
    routing.AddDimension(
        time_evaluator_index,
        horizon,  # allow waiting time
        horizon,  # maximum time per vehicle
        False,  # don't force start cumul to zero since we are giving TW to start nodes
        time)
    time_dimension = routing.GetDimensionOrDie(time)
    # Add time window constraints for each location except depot
    # and 'copy' the slack var in the solution object (aka Assignment) to print it
    for location_idx, time_window in enumerate(data['time_windows']):
        if location_idx == 0:
            continue
        index = manager.NodeToIndex(location_idx)
        time_dimension.CumulVar(index).SetRange(time_window[0], time_window[1])
        routing.AddToAssignment(time_dimension.SlackVar(index))
    # Add time window constraints for each vehicle start node
    # and 'copy' the slack var in the solution object (aka Assignment) to print it
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        time_dimension.CumulVar(index).SetRange(data['time_windows'][0][0],
                                                data['time_windows'][0][1])
        routing.AddToAssignment(time_dimension.SlackVar(index))
        # Warning: Slack var is not defined for vehicle's end node
        #routing.AddToAssignment(time_dimension.SlackVar(self.routing.End(vehicle_id)))


###########
# Printer #
###########

# change the arrive earliest to depotTime + the specified time in minutes.
def print_solution(data, manager, routing, assignment):  # pylint:disable=too-many-locals
    """Prints assignment on console"""
    print('Objective: {}'.format(assignment.ObjectiveValue()))
    total_distance = 0
    total_load = 0
    total_time = 0
    capacity_dimension = routing.GetDimensionOrDie('Capacity')
    time_dimension = routing.GetDimensionOrDie('Time')
    
#     saves the routing data
    run_id = []
    record_id = []
    pickup_zip = []
    arrive_earliest = []
    arrive_latest = []
    destination_zip = []
    depart_earliest = []
    depart_latest = []
    arrival_if_leave_now = []
    truck_id = []
    truck_type = []
    unit_count = []
    temperature = []
    cost = []
    c = create_time_evaluator(data)
#         prints the output into a dataframe
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        plan_output = 'Route for vehicle {}:\n'.format(vehicle_id)
        distance = 0
        while not routing.IsEnd(index):
            load_var = capacity_dimension.CumulVar(index)
            time_var = time_dimension.CumulVar(index)
            slack_var = time_dimension.SlackVar(index)

            run_id.append(len(run_id) + 1)
            record_id.append(bananasZips['record_id'][manager.IndexToNode(index)])
            pickup_zip.append(bananasZips['shipper_zip'][manager.IndexToNode(index)])
            arrive_earliest.append(depotTime + timedelta(minutes = assignment.Min(time_var)))
            arrive_latest.append(depotTime + timedelta(minutes = assignment.Max(time_var)))
            destination_zip.append(bananasZips['shipper_zip'][manager.IndexToNode(assignment.Value(routing.NextVar(index)))])
            depart_earliest.append(depotTime + timedelta(minutes = assignment.Min(time_var) + assignment.Min(slack_var)))
            depart_latest.append(depotTime + timedelta(minutes = assignment.Min(time_var) + assignment.Max(slack_var)))
            arrival_if_leave_now.append(depart_earliest[len(depart_earliest) - 1] + timedelta(minutes = data['time_matrix'][manager.IndexToNode(index)][manager.IndexToNode(assignment.Value(routing.NextVar(index)))]))
            truck_id.append(vehicle_id)
            if (vehicle_id + 1  <= data['num_16trucks']):
                truck_type.append("16'")
            else:
                truck_type.append("53'")
            unit_count.append(bananasZips['unit_count'][manager.IndexToNode(index)])
            temperature.append(temperature_filter)
            cost.append(routing.GetArcCostForVehicle(index, assignment.Value(routing.NextVar(index)), vehicle_id))
            plan_output +='{0} Load({1}) Time({2},{3}) Slack({4},{5}) Cost({6})-> {7}\n'.format(
            manager.IndexToNode(index),
            assignment.Value(load_var),
            assignment.Min(time_var),
            assignment.Max(time_var),
            assignment.Min(slack_var), assignment.Max(slack_var),
            routing.GetArcCostForVehicle(index, assignment.Value(routing.NextVar(index)),
                                                 vehicle_id),
            manager.IndexToNode(assignment.Value(routing.NextVar(index))))
            previous_index = index
            index = assignment.Value(routing.NextVar(index))
            distance += routing.GetArcCostForVehicle(previous_index, index,
                                                     vehicle_id)
        load_var = capacity_dimension.CumulVar(index)
        time_var = time_dimension.CumulVar(index)
        slack_var = time_dimension.SlackVar(index)
        plan_output += '{0} Load({1}) Time({2},{3})\n'.format(
            manager.IndexToNode(index),
            assignment.Value(load_var),
            assignment.Min(time_var), assignment.Max(time_var))
        plan_output += 'Cost of the route: {0}m\n'.format(distance)
        plan_output += 'Load of the route: {}\n'.format(
            assignment.Value(load_var))
        plan_output += 'Time of the route: {}\n'.format(
            assignment.Value(time_var))
        print(plan_output)
        total_distance += distance
        total_load += assignment.Value(load_var)
        total_time += assignment.Value(time_var)
    print('Total Cost of all routes: {0}m'.format(total_distance))
    print('Total Load of all routes: {}'.format(total_load))
    print('Total Time of all routes: {0}min'.format(total_time))
    d = {'run_id':run_id, 'record_id': record_id, 'arrive_earliest' : pd.to_datetime(arrive_earliest), 
         'arrive_latest' : pd.to_datetime(arrive_latest), 'pickup_zip' : pickup_zip, 'depart_earliest' : pd.to_datetime(depart_earliest),
         'depart_latest' : pd.to_datetime(depart_latest), 'arrival_if_leave_now': arrival_if_leave_now,  'destination_zip' : destination_zip, 'truck_id' : truck_id, 'truck_type' : truck_type, 'unit_count' : unit_count, 'temperature' : temperature,
         'cost' : cost}
    output = pd.DataFrame(data=d)
    return output

########
# Main #
########
"""Entry point of the program"""
# Instantiate the data problem.
data = create_data_model()

# Create the routing index manager
manager = pywrapcp.RoutingIndexManager(len(data['demands']),
                                       data['num_vehicles'], data['depot'])

# Create Routing Model
routing = pywrapcp.RoutingModel(manager)

# Define weight of each edge of the graph
for i in range(data['num_vehicles']):
    cost_evaluator_index = routing.RegisterTransitCallback(
        partial(create_cost_evaluator(data, i), manager))
    routing.SetArcCostEvaluatorOfVehicle(cost_evaluator_index, i)

# Add Capacity constraint
demand_evaluator_index = routing.RegisterUnaryTransitCallback(
    partial(create_demand_evaluator(data), manager))
add_capacity_constraints(routing, data, demand_evaluator_index)

# Add Time Window constraint
time_evaluator_index = routing.RegisterTransitCallback(
    partial(create_time_evaluator(data), manager))
add_time_window_constraints(routing, manager, data, time_evaluator_index)

# Setting first solution heuristic (cheapest addition).
search_parameters = pywrapcp.DefaultRoutingSearchParameters()
search_parameters.first_solution_strategy = (
    routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)  # pylint: disable=no-member
# Solve the problem.
assignment = routing.SolveWithParameters(search_parameters)
output = print_solution(data, manager, routing, assignment)
```

```python
# hub to receiving sites solver
from functools import partial

from ortools.constraint_solver import pywrapcp
from ortools.constraint_solver import routing_enums_pb2

###########################
# Problem Data Definition #
###########################
def create_data_model():
    """Stores the data for the problem"""

    data = {}
    data['time_matrix'] = rs_time_matrix
    data['time_windows'] = rs_time_windows
    data['demands'] = rs_zips['unit_count']
    data['time_per_demand_unit'] = 0 # 0 minutes/unit
    data['temperature'] = temperature_filter
#     TODO: Change number of trucks
    data['num_16trucks'] = 5
    data['num_53trucks'] = 10
    data['num_vehicles'] = data['num_16trucks'] + data['num_53trucks']
    data['vehicle_capacities'] = [unit_capacity_16]*data['num_16trucks'] + [unit_capacity_53]*data['num_53trucks']
    data['depot'] = 0
    data['output'] = []
    return data


#######################
# Problem Constraints #
#######################

def create_demand_evaluator(data):
    """Creates callback to get demands at each location."""
    _demands = data['demands']

    def demand_evaluator(manager, node):
        """Returns the demand of the current node"""
        return _demands[manager.IndexToNode(node)]

    return demand_evaluator


def create_time_evaluator(data):
    def travel_time(data, from_node, to_node):
        """Gets the travel times between two locations."""
        if from_node == to_node:
            travel_time = 0
        else:
            travel_time = data['time_matrix'][from_node][to_node]
        return travel_time

    _total_time = data['time_matrix']
    def time_evaluator(manager, from_node, to_node):
        """Returns the total time between the two nodes"""
        return _total_time[manager.IndexToNode(from_node)][manager.IndexToNode(
            to_node)]
    return time_evaluator

# cost evaluator per vehicle
def create_cost_evaluator(data,vehindex):
    """Created call back to get total cost between locations """
    def travel_cost(data, from_node, to_node):
        if from_node == to_node:
            return 0
        else:
            hrs = (float)(data['time_matrix'][from_node][to_node] / 60)
            
            if (data['vehicle_capacities'][vehindex] == unit_capacity_16):
                rent = rent_16
                labor = labor_16
                mpg = mpg_16
                temp_cost = b_info['temp_cost_16']
            elif (data['vehicle_capacities'][vehindex] == unit_capacity_53):
                rent = rent_53
                labor = labor_53
                mpg = mpg_53
                temp_cost = b_info['temp_cost_53']
            cost = (rent + labor + temp_cost) * hrs + gas_price * hrs * truck_speed/mpg
            return cost
        
    _total_cost = [[0]*len(data['demands']) for i in range(len(data['demands']))]
    for from_node in range(len(data['demands'])):
        _total_cost[from_node] = {}
        for to_node in range(len(data['demands'])):
            if from_node == to_node:
                _total_cost[from_node][to_node] = 0
            else:
                _total_cost[from_node][to_node] = (travel_cost(data, from_node, to_node))
    def cost_evaluator(manager, from_node, to_node):
        return _total_cost[manager.IndexToNode(from_node)][manager.IndexToNode(
            to_node)]
    return cost_evaluator

def add_capacity_constraints(routing, data, demand_evaluator_index):
    """Adds capacity constraint"""
    capacity = 'Capacity'
    routing.AddDimensionWithVehicleCapacity(
        demand_evaluator_index,
        0,  # null capacity slack
        data['vehicle_capacities'],
        True,  # start cumul to zero
        capacity)

def add_time_window_constraints(routing, manager, data, time_evaluator_index):
    """Add Global Span constraint"""
    time = 'Time'
    horizon = 12000000
    routing.AddDimension(
        time_evaluator_index,
        horizon,  # allow waiting time
        horizon,  # maximum time per vehicle
        False,  # don't force start cumul to zero since we are giving TW to start nodes
        time)
    time_dimension = routing.GetDimensionOrDie(time)
    # Add time window constraints for each location except depot
    # and 'copy' the slack var in the solution object (aka Assignment) to print it
    for location_idx, time_window in enumerate(data['time_windows']):
        if location_idx == 0:
            continue
        index = manager.NodeToIndex(location_idx)
        time_dimension.CumulVar(index).SetRange(time_window[0], time_window[1])
        routing.AddToAssignment(time_dimension.SlackVar(index))
    # Add time window constraints for each vehicle start node
    # and 'copy' the slack var in the solution object (aka Assignment) to print it
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        time_dimension.CumulVar(index).SetRange(data['time_windows'][0][0],
                                                data['time_windows'][0][1])
        routing.AddToAssignment(time_dimension.SlackVar(index))
        # Warning: Slack var is not defined for vehicle's end node
        #routing.AddToAssignment(time_dimension.SlackVar(self.routing.End(vehicle_id)))


###########
# Printer #
###########

# change the arrive earliest to depotTime + the specified time in minutes.
def print_solution(data, manager, routing, assignment):  # pylint:disable=too-many-locals
    """Prints assignment on console"""
    print('Objective: {}'.format(assignment.ObjectiveValue()))
    total_distance = 0
    total_load = 0
    total_time = 0
    capacity_dimension = routing.GetDimensionOrDie('Capacity')
    time_dimension = routing.GetDimensionOrDie('Time')
    
    run_id = []
    record_id = []
    drop_off_zip = []
    arrive_earliest = []
    arrive_latest = []
    destination_zip = []
    depart_earliest = []
    depart_latest = []
    arrival_if_leave_now = []
    truck_id = []
    truck_type = []
    unit_count = []
    temperature = []
    cost = []
    
    c = create_time_evaluator(data)
    
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        plan_output = 'Route for vehicle {}:\n'.format(vehicle_id)
        distance = 0
        while not routing.IsEnd(index):
            load_var = capacity_dimension.CumulVar(index)
            time_var = time_dimension.CumulVar(index)
            slack_var = time_dimension.SlackVar(index)

            run_id.append(len(run_id) + 1)
            record_id.append(rs_zips['record_id'][manager.IndexToNode(index)])
            drop_off_zip.append(rs_zips['destination_zip'][manager.IndexToNode(index)])
            arrive_earliest.append(rs_start_time + timedelta(minutes = assignment.Min(time_var)))
            arrive_latest.append(rs_start_time + timedelta(minutes = assignment.Max(time_var)))
            destination_zip.append(rs_zips['destination_zip'][manager.IndexToNode(assignment.Value(routing.NextVar(index)))])
            depart_earliest.append(rs_start_time + timedelta(minutes = assignment.Min(time_var) + assignment.Min(slack_var)))
            depart_latest.append(rs_start_time + timedelta(minutes = assignment.Min(time_var) + assignment.Max(slack_var)))
            arrival_if_leave_now.append(depart_earliest[len(depart_earliest) - 1] + timedelta(minutes = data['time_matrix'][manager.IndexToNode(index)][manager.IndexToNode(assignment.Value(routing.NextVar(index)))]))
            truck_id.append(vehicle_id)
            if (vehicle_id + 1  <= data['num_16trucks']):
                truck_type.append("16'")
            else:
                truck_type.append("53'")
            unit_count.append(rs_zips['unit_count'][manager.IndexToNode(index)])
            temperature.append(temperature_filter)
            cost.append(routing.GetArcCostForVehicle(index, assignment.Value(routing.NextVar(index)), vehicle_id))
            plan_output +='{0} Load({1}) Time({2},{3}) Slack({4},{5}) Cost({6})-> {7}\n'.format(
            manager.IndexToNode(index),
            assignment.Value(load_var),
            assignment.Min(time_var),
            assignment.Max(time_var),
            assignment.Min(slack_var), assignment.Max(slack_var),
            routing.GetArcCostForVehicle(index, assignment.Value(routing.NextVar(index)),
                                                 vehicle_id),
            manager.IndexToNode(assignment.Value(routing.NextVar(index))))
            previous_index = index
            index = assignment.Value(routing.NextVar(index))
            distance += routing.GetArcCostForVehicle(previous_index, index,
                                                     vehicle_id)
        load_var = capacity_dimension.CumulVar(index)
        time_var = time_dimension.CumulVar(index)
        slack_var = time_dimension.SlackVar(index)
        plan_output += '{0} Load({1}) Time({2},{3})\n'.format(
            manager.IndexToNode(index),
            assignment.Value(load_var),
            assignment.Min(time_var), assignment.Max(time_var))
        plan_output += 'Cost of the route: {0}m\n'.format(distance)
        plan_output += 'Load of the route: {}\n'.format(
            assignment.Value(load_var))
        plan_output += 'Time of the route: {}\n'.format(
            assignment.Value(time_var))
        print(plan_output)
        total_distance += distance
        total_load += assignment.Value(load_var)
        total_time += assignment.Value(time_var)
    print('Total Cost of all routes: {0}m'.format(total_distance))
    print('Total Load of all routes: {}'.format(total_load))
    print('Total Time of all routes: {0}min'.format(total_time))
    d = {'run_id':run_id, 'record_id': record_id, 'arrive_earliest' : pd.to_datetime(arrive_earliest), 
         'arrive_latest' : pd.to_datetime(arrive_latest), 'drop_off_zip' : drop_off_zip, 'depart_earliest' : pd.to_datetime(depart_earliest),
         'depart_latest' : pd.to_datetime(depart_latest), 'arrival_if_leave_now': arrival_if_leave_now,  'destination_zip' : destination_zip, 'truck_id' : truck_id, 'truck_type' : truck_type, 'unit_count' : unit_count, 'temperature' : temperature,
         'cost' : cost}
    output = pd.DataFrame(data=d)
    return output

########
# Main #
########
"""Entry point of the program"""
# Instantiate the data problem.
data = create_data_model()

# Create the routing index manager
manager = pywrapcp.RoutingIndexManager(len(data['demands']),
                                       data['num_vehicles'], data['depot'])

# Create Routing Model
routing = pywrapcp.RoutingModel(manager)

# Define weight of each edge
for i in range(data['num_vehicles']):
    cost_evaluator_index = routing.RegisterTransitCallback(
        partial(create_cost_evaluator(data, i), manager))
    routing.SetArcCostEvaluatorOfVehicle(cost_evaluator_index, i)

# Add Capacity constraint
demand_evaluator_index = routing.RegisterUnaryTransitCallback(
    partial(create_demand_evaluator(data), manager))
add_capacity_constraints(routing, data, demand_evaluator_index)

# Add Time Window constraint
time_evaluator_index = routing.RegisterTransitCallback(
    partial(create_time_evaluator(data), manager))
add_time_window_constraints(routing, manager, data, time_evaluator_index)

# Setting first solution heuristic (cheapest addition).
search_parameters = pywrapcp.DefaultRoutingSearchParameters()
search_parameters.first_solution_strategy = (
    routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)  # pylint: disable=no-member
# Solve the problem.
assignment = routing.SolveWithParameters(search_parameters)
rs_output = print_solution(data, manager, routing, assignment)
```
